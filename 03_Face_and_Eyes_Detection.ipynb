### Import Libraries

```python
import numpy as np
import cv2 
import matplotlib.pyplot as plt
%matplotlib inline 
```

### Create Classifiers

```python
face_classifier = cv2.CascadeClassifier("Haarcascades/Haarcascade_frontalface_default.xml")
eye_classifier = cv2.CascadeClassifier("Haarcascades/haarcascade_eye.xml")
```

### Open Image

```python
img = cv2.imread("Images/eye_face.jpg")
```

### Fix Image

```python
fix_image = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(fix_image)
```

### Face Classifier

```python
faces = face_classifier.detectMultiScale(fix_image, 1.3, 5)
```

### If No Face Detected

```python
if faces is ():
    print("no Faces Found")
```

### It's a Kind of Magic

```python
def detect_faces_eyes(fix_image):
    face_rects = face_classifier.detectMultiScale(fix_image)
    
    for (x,y,w,h) in face_rects:
        cv2.rectangle(fix_image , 
                     (x,y),
                     (x+w , y+h),
                     (255,0,0),
                     7)
    eyes_rects = eye_classifier.detectMultiScale(fix_image)
    for (ix,iy,iw,ih) in eyes_rects:
                cv2.rectangle(fix_image , 
                     (ix,iy),
                     (ix+iw , iy+ih),
                     (0,0,255),
                     5)
    return fix_image   
```

```python
result = detect_faces_eyes(fix_image)
plt.imshow(result)
```
